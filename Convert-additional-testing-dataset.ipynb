```python
import os
```

```python
def collect_image_files(directory):
    image_files = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.jpg') or file.endswith('.png') or file.endswith('.ppm'):
                image_files.append(os.path.join(root, file))
    return image_files
```

```python
all_test_image = collect_image_files('./trafficsigns_dataset_testing')
all_test_image
```

```python
from PIL import Image

def resize_images_in_place(image_paths, target_size=(28, 28)):
    for image_path in image_paths:
        # Open the image file
        image = Image.open(image_path)
        
        # Resize the image
        resized_image = image.resize(target_size, Image.ANTIALIAS)
        
        # Overwrite the original image with the resized image
        resized_image.save(image_path)


target_size = (28, 28)

# Resize images in place (overwrite original images)
resize_images_in_place(all_test_image, target_size)
```

```python
def convert_ppm_to_png_in_place(image_paths):
    for image_path in image_paths:
        # Open the PPM image
        ppm_image = Image.open(image_path)
        
        # Convert to RGB mode if necessary (some PPM images may be in RGB mode)
        if ppm_image.mode != 'RGB':
            ppm_image = ppm_image.convert('RGB')
        
        # Extract the filename (without extension) from the image path
        filename = os.path.splitext(os.path.basename(image_path))[0]
        
        # Specify the output PNG file path (overwrite the original PPM image)
        output_path = os.path.join(os.path.dirname(image_path), f"{filename}.png")
        
        # Save the image as PNG format (overwriting the original file)
        ppm_image.save(output_path, format='PNG')
        
        # Delete the original PPM image file
        os.remove(image_path)
        print(f"Converted and deleted: {image_path}")

```

```python
convert_ppm_to_png_in_place(all_test_image)
```
